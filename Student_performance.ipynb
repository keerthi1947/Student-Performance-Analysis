# Load Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('student-mat.csv', sep=';')
print('Data loaded Successfully!!')
```

# Explore Dataset

```python
print(df.head())
print("Shape of dataset:", df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

# Remove Duplicates

```python
df = df.drop_duplicates()
print("Duplicates removed")
```

# Data Analysis

```python
average_grade = df['G3'].mean()
print("Average Final Grade:", average_grade)
high_scores = df[df['G3'] > 15]
print("Number of students scoring above 15:", len(high_scores))
correlation = df['studytime'].corr(df['G3'])
print("Correlation between study time and grades:", correlation)
gender_avg = df.groupby('sex')['G3'].mean()
print(gender_avg)
```

 # Data Visualization

```python
plt.figure(figsize=(8,5))

plt.hist(df['G3'], bins=10)

plt.title('Distribution of Final Grades')
plt.xlabel('Final Grade')
plt.ylabel('Number of Students')

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x='studytime', y='G3', data=df)

plt.title('Study Time vs Final Grade')

plt.show()
```

```python
plt.figure(figsize=(6,5))

gender_avg.plot(kind='bar')

plt.title('Average Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Final Grade')

plt.show()
```

# Conclusion

```python
print('- The dataset was cleaned and analyzed successfully.')
print('- Average student performance was calculated using G3 scores.')
print('- Students with higher study time generally performed better.')
print('- Gender-wise performance comparison was analyzed.')
print('- Visualizations helped understand grade distribution and study patterns.')
```
